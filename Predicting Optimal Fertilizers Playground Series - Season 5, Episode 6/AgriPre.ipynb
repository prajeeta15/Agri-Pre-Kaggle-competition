```python
train = pd.read_csv('/kaggle/input/playground-series-s5e6/train.csv')
          original = pd.read_csv('/kaggle/input/playground-series-s5e6/train.csv')
          test = pd.read_csv('/kaggle/input/playground-series-s5e6/test.csv')
          submission = pd.read_csv('/kaggle/input/playground-series-s5e6/sample_submission.csv')
```

```python
for df in (train, test, original):
    if 'id' in df.columns:
        df.drop('id', axis=1, inplace=True)

train = pd.concat([train, original], ignore_index=True)
train.drop_duplicates(inplace=True)
```

```python
cat_cols = [col for col in train.select_dtypes(include=['object', 'category']).columns if col != "Fertilizer Name"]
for col in cat_cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])
```

```python
# Label encode categorical columns
label_encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])
    label_encoders[col] = le

# Encode target
fer_label_enc = LabelEncoder()
train["Fertilizer Name"] = fer_label_enc.fit_transform(train["Fertilizer Name"])
```

```python
X = train.drop(columns=["Fertilizer Name"])
y = train["Fertilizer Name"]
X_test = test
```

```python
model = lgb.LGBMClassifier(device='gpu',
    boosting_type='gbdt',
    objective='multiclass',
    num_class=len(np.unique(y)),
    metric='multi_logloss',
    random_state=42)
```

```python
def mapk(actual, predicted, k=3):
    def apk(a, p, k):
        p = p[:k]
        score = 0.0
        hits = 0.0
        seen = set()
        for i, pred in enumerate(p):
            if pred in a and pred not in seen:
                hits += 1.0
                score += hits / (i + 1.0)
                seen.add(pred)
        return score / min(len(a), k)
    actual_wrapped = [[a] for a in actual]
    return np.mean([apk(a, p, k) for a, p in zip(actual_wrapped, predicted)])
```

```python
def objective(trial):
    params = {
        'objective': 'multiclass',
        'num_class': len(np.unique(y)),
        'boosting_type': 'gbdt',
        'metric': 'multi_logloss',
        'learning_rate': trial.suggest_float("learning_rate", 0.01, 0.1),
        'max_depth': trial.suggest_int("max_depth", 3, 10),
        'num_leaves': trial.suggest_int("num_leaves", 15, 150),
        'colsample_bytree': trial.suggest_float("colsample_bytree", 0.3, 0.9),
        'subsample': trial.suggest_float("subsample", 0.5, 1.0),
        'device_type': 'cpu', # Keep this as cpu for Optuna trials unless you have specific GPU support configured
        'n_estimators': 1000,
        'early_stopping_rounds': 30,
        'verbose': -1
    }

    skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    map3_scores = []
    # Use X.values and y.values for splitting with numpy array

    for train_idx, val_idx in skf.split(X.values, y.values):
        y_train_fold = y.iloc[train_idx].values
        y_val_fold = y.iloc[val_idx].values

    # Skip folds with unseen labels
        if set(y_val_fold) - set(y_train_fold):
           continue

        model = lgb.LGBMClassifier(**params)
        model.fit(X.iloc[train_idx], y.iloc[train_idx],
              eval_set=[(X.iloc[val_idx], y.iloc[val_idx])],
              eval_metric='multi_logloss',
              callbacks=[lgb.early_stopping(params['early_stopping_rounds'])])

        preds = model.predict_proba(X.iloc[val_idx])
        top3 = np.argsort(preds, axis=1)[:, -3:][:, ::-1]
        map3_scores.append(mapk(y_val_fold, top3))

    return np.mean(map3_scores)


# %%
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50, timeout=3000)
```

```python
print("Best trial:")
print(f"  Value: {study.best_value}")
print("  Params:")
for key, value in study.best_params.items():
    print(f"    {key}: {value}")
```

```python
%%time

FOLDS = 5
skf = StratifiedKFold(n_splits=FOLDS, shuffle=True, random_state=42)

num_classes = y.nunique()
oof = np.zeros((len(train), num_classes))
pred_prob = np.zeros((len(test), num_classes))

# Define LGBM model
lgb_params = {
    'objective': 'multiclass',
    'num_class': len(np.unique(y)),
    'boosting_type': 'gbdt',
    'metric': 'multi_logloss',
    'max_depth': 5,
    'learning_rate': 0.05,
    'subsample': 0.8,
    'max_bin': 128,
    'colsample_bytree': 0.6,
    'random_state': 42,
    'n_estimators': 10000,
    'early_stopping_rounds': 50,
    'verbose': 100
}

for fold, (train_idx, valid_idx) in enumerate(skf.split(X, y), 1):
    print(f"{'='*10} Fold {fold} {'='*10}")

    X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
    y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

    # Skip fold if validation labels not all in training labels
    if set(y_valid) - set(y_train):
        print(f"❌ Skipping Fold {fold} — unseen labels in validation set")
        continue

    model = lgb.LGBMClassifier(**lgb_params)
    model.fit(X_train, y_train, eval_set=[(X_valid, y_valid.values)])

    oof[valid_idx] = model.predict_proba(X_valid)
    pred_prob += model.predict_proba(X_test) / FOLDS

    top_3_preds = np.argsort(oof[valid_idx], axis=1)[:, -3:][:, ::-1]
    actual = [[label] for label in y_valid]
    map3_score = mapk(actual, top_3_preds)
    print(f"✅ Fold {fold} MAP@3 Score: {map3_score:.5f}")


# Import log_loss
from sklearn.metrics import log_loss

logloss_score = log_loss(y, oof)
print(f"\n📊 Overall Log Loss: {logloss_score:.5f}")
```

```python
# Get top 3 predicted class indices
top_3_preds = np.argsort(pred_prob, axis=1)[:, -3:][:, ::-1]  # shape: (n_samples, 3)

# Decode class indices back to original fertilizer names
top_3_labels = fer_label_enc.inverse_transform(top_3_preds.ravel()).reshape(top_3_preds.shape)

# Reload the test data specifically for the submission file to get the 'id' column
test_for_submission = pd.read_csv('/content/test.csv')

submission = pd.DataFrame({
    'id': test_for_submission['id'], # Use the reloaded test data for 'id'
    'Fertilizer Name': [' '.join(row) for row in top_3_labels]
})

# Save to CSV
submission.to_csv('submission.csv', index=False)
print("✅ Submission file saved as 'submission.csv'")
```
